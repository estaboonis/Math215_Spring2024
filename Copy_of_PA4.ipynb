## Problem 1 (12 points)  

Complete the implementation of a `Circle` class.

*  Constructor:  `def __init__(self, center_x = 0, center_y = 0, radius = 0):`

*  String method:  `def __str__(self)`

*  Area:  `def area(self)`

*  Contains:  `def contains(self, other)`:

*  Intersects:  `def intersects(self, other)`:

*  Less Than:  `def __lt__(self, other):`  Compare based on the circle's radius



```python
import math

class Circle:
    def __init__(self, center_x = 0, center_y = 0, radius = 0):
        self.center_x = center_x
        self.center_y = center_y
        self.radius = radius

    def __str__(self):
        return f'this is a circle of origin ({self.center_x},{self.center_y}) and radius {self.radius}'

    def area(self):
        return math.pi *(self.radius * self.radius)

    def intersects(self, other):
        if (self.radius + other.radius) <= math.sqrt(((other.center_x - self.center_x)**2)+((other.center_y - self.center_y)**2)):
            if ((math.sqrt(((other.center_x - self.center_x)**2)+((other.center_y - self.center_y)**2)) + other.radius) < self.radius) or ((math.sqrt(((other.center_x - self.center_x)**2)+((other.center_y - self.center_y)**2)) + self.radius) < other.radius):
                return False
            else:
                return True
        else:
            return False

    def contains(self, other):
        if ((math.sqrt(((other.center_x - self.center_x)**2)+((other.center_y - self.center_y)**2)) + other.radius) < self.radius) or ((math.sqrt(((other.center_x - self.center_x)**2)+((other.center_y - self.center_y)**2)) + self.radius) < other.radius):
            return True
        else:
            return False

    def __lt__(self, other):
        if self.radius < other.radius:
            return True

```
